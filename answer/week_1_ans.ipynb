10. Implement the Game of Life using numpy arrays

Idea:

The most important step is to combute the number of living neighboring cells for each cell. Instead using for loops, a faster implementation takes advantage of numpy.

We denote 0 for dead cells and 1 for living cells. Suppose `Z` is the board for the Game of Life with shape $(m+2,n+2)$, where all edge cells are 0. That is, the actual board is `Z[1:-1,1:-1]` of shape $(m,n)$. If we have living edge cells, we can use the technique in Question 2 to create such a board `Z`. 

A fast way to count the living neighbors is to do "vectorize" the computation. The number of living neighbors is the sum of the status of 

upper-left neighbor, upper neighbor, upper-right neighbor,

left neighbor,  $\hspace{55mm}$                      right neighbor, 

lower-left neighbor, lower neighbor, and lower-right neighbor.

The status of upper-left neighbor for each cell is exactly `Z[0:-2,0:-2]`; (the upper-left neighbor of `Z[i,j]` is `Z[i-1,j-1]`)

The status of upper neighbor for each cell is exactly `Z[0:-2,1:-1]`;

$\dots$

The status of lower-right neighbor for each cell is exactly `Z[2:  ,2:]`.

Taking the sum of above 8 status blocks, we get the number of living neighbors for each cell on the actual board `Z[1:-1,1:-1]`.

Below is a sample answer that has the fastest speed among our answers.



```python
# Author: Nicolas Rougier

def iterate(Z):
    # Count neighbours
    N = (Z[0:-2,0:-2] + Z[0:-2,1:-1] + Z[0:-2,2:] +
         Z[1:-1,0:-2]                + Z[1:-1,2:] +
         Z[2:  ,0:-2] + Z[2:  ,1:-1] + Z[2:  ,2:])

    # Apply rules
    birth = (N==3) & (Z[1:-1,1:-1]==0)
    survive = ((N==2) | (N==3)) & (Z[1:-1,1:-1]==1)
    Z[...] = 0
    Z[1:-1,1:-1][birth | survive] = 1
    return Z

Z = np.random.randint(0,2,(50,50))
for i in range(100): Z = iterate(Z)
print(Z)
```

6. Which two species have the most similar mean weight?

```python
info = df_full.groupby('species')['body_mass_g'].mean().to_dict()
print(f'info = {info}')

diff = np.infty
arg_x = arg_y = None
for x, wx in info.items():
  for y, wy in info.items():
    if x == y: continue

    curr_diff = np.abs(wx - wy)
    if curr_diff < diff:
      arg_x, arg_y = x , y
      diff = curr_diff

print(f'species {arg_x} and {arg_y} have the most similar mean weight.')
```

Adelie and Chinstrap have the most similar mean weight.

10. From the dataframe in Question 9, take the mean of the numeric columns, grouped at one hour intervals.  Save the result as df_means, and display it.

```python
# Method 1
taxis_one_day.groupby([taxis_one_day["pickup_time"].dt.hour]).mean()
```

```python
# Method 2
taxis_one_day.groupby(taxis_one_day.pickup_time.map(lambda x: x.hour)).mean()
```

```python
# Method 3
taxis_one_day.groupby(pd.Grouper(key='pickup_time', freq='60Min')).mean()
```

```python
# Method 4
taxis_one_day.set_index('pickup_time').resample('60min').mean()
```
